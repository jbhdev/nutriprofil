# Analyse des Relations entre Variables Nutritionnelles

À partir de la matrice de scatterplots, voici les principales observations sur les paires de variables :

---

## 1. Calories vs. Glucides  
- **Légère corrélation positive** : en général, plus un aliment est riche en glucides, plus il est calorique.  
- **Dispersion large** : certains aliments très riches en graisses ou en protéines (fruits oléagineux, fromages) ont beaucoup de calories malgré peu de glucides.

---

## 2. Calories vs. Gras  
- **Corrélation forte** : la pente ascendante claire montre que les aliments gras (huiles, noix) sont très caloriques.  
- **Effet « double mode »** : on distingue deux nuages :  
  - aliments riches en graisses (> 20 g) avec calories > 200 kcal,  
  - aliments à faible teneur en gras (0–10 g) pour la plupart < 200 kcal.  

---

## 3. Calories vs. Protéines  
- **Corrélation modérée** : les protéines contribuent aux calories, mais l’effet est moins marqué que pour les lipides.  
- **Nuage diffus** : on voit quelques aliments très protéinés (viandes, fromages) à 15–40 g de protéines pour 200–400 kcal.

---

## 4. Glucides vs. CG (Charge Glycémique)  
- **Corrélation très forte** : la CG augmente presque linéairement avec les glucides.  
- **Nuance IG** : la dispersion autour de la droite est due aux différences d’indice glycémique (IG) — deux aliments à mêmes glucides peuvent avoir CG différentes si leur IG diffère.

---

## 5. IG vs. CG  
- **Corrélation positive** : pour un même taux de glucides, plus l’IG est élevé, plus la CG l’est aussi.  
- **Nuage triangulaire** :  
  - bas IG (< 30) → CG généralement basse,  
  - haut IG (> 70) → CG très variable selon la quantité de glucides.

---

## 6. Fibres vs. CG  
- **Tendance légèrement négative** : beaucoup d’aliments riches en fibres ont une CG modérée ou faible, car les fibres ralentissent l’absorption des glucides.  
- **Points isolés** : quelques aliments très fibreux (légumineuses, céréales complètes) conservent une CG modérée car ils contiennent aussi des glucides digestibles.

---

## 7. Gras vs. CG  
- **Nuage diffus** : pas de lien direct fort entre lipides et CG, ce qui est logique car la CG dépend surtout des glucides et de l’IG.  
- **Bruit de fond** : aliments très gras (huiles, fruits oléagineux) ont CG proche de zéro, car ils contiennent peu ou pas de glucides.

---

## 8. Protéines vs. CG  
- **Aucune corrélation marquée** : la CG n’est pas influencée par la teneur en protéines, d’où un nuage presque aléatoire.  
- **Exceptions** : aliments très protéinés et peu glucidiques (viandes, poissons) ont CG = 0, formant une bande horizontale près de zéro.

---

### Conclusion  
- **CG** est avant tout déterminée par les **glucides** et l’**indice glycémique**.  
- **Calories** sont fortement liées aux **graisses**, puis aux **glucides**, enfin aux **protéines**.  
- Les **fibres** jouent un rôle protecteur : leur présence modère légèrement la CG.  
- Pour la modélisation, il conviendra de capturer ces interactions (e.g. glucides × IG) et de gérer la non-linéarité de certaines relations (calories vs. graisses).

  - plats légers (200–300 kcal) CG 0–20,  
  - plats riches (300–500 kcal) CG 20–40.

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Suppose que df est ton DataFrame avec les variables quantitatives uniquement
# Tu peux ajuster la sélection selon ton besoin
X = df[['Calories', 'Fibres', 'Glucides', 'Gras', 'IG', 'Proteines']]  # Retire CG si c'est la variable cible

# Ajoute une constante (intercept) nécessaire au calcul du VIF
X = add_constant(X)

# Calcul du VIF
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Affichage du tableau VIF
print(vif_data)
```

# Analyse des VIF (Variance Inflation Factor)

Le **VIF** mesure la **colinéarité** entre variables explicatives :  
- Un VIF > 10 signale une multicolinéarité élevée.  
- Entre 5 et 10, on peut s’interroger sur un possible chevauchement d’information.  
- < 5, la colinéarité est généralement considérée comme acceptable.
