```python
import pandas as pd
import numpy as np


df_conso_ménages = pd.read_csv("/Users/coulibalykani/Desktop/Projet 3/nutriprofil/Dataframe/conso-menages-2024.csv", sep=',')
df_conso_ménages
```

```python
#Conversion des colonnes de consommation en float

cols_annees = ['2010', '2011', '2012', '2013', '2014', '2015', '2016',
               '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']

for col in cols_annees:
    df_conso_ménages[col] = df_conso_ménages[col].astype(str)             # convertir en string
    df_conso_ménages[col] = df_conso_ménages[col].str.replace(r'\s+', '', regex=True)  # enlever tous les espaces
    df_conso_ménages[col] = df_conso_ménages[col].str.replace(',', '.')   # (optionnel : convertir ',' en '.' si besoin)
    df_conso_ménages[col] = df_conso_ménages[col].astype(float)           # convertir en float
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Sélection des caractéristiques et de la cible
Features_conso = df_conso_ménages[['2010', '2011', '2012', '2013', '2014', '2015', 
                                   '2016', '2017', '2018', '2019', '2020', '2021', 
                                   '2022', '2023']]

Target_conso = df_conso_ménages['2024']

X_train,X_test,y_train,y_test = train_test_split(Features_conso, Target_conso, test_size=0.2, random_state=42)  
```

```python
# Entraînement du modèle

model_conso = Lasso()
model_conso.fit(X_train, y_train)   
```

```python
# Prédiction du modèle

y_pred = model_conso.predict(X_test)    
```

```python
# R2 et RMSE

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))  
print(f"R2: {r2}")
print(f"RMSE: {rmse}")
```

```python
# Importation du fichier csv en Data Frame

df= pd.read_csv('/Users/coulibalykani/Desktop/Projet 3/nutriprofil/Dataframe/dataframe_complet_rempli_proteines.csv')
# Remplacement de/Users/coulibalykani/Desktop/Projet 3/nutriprofil/Dataframe/dataframe_complet_rempli_proteines.csv
df['Famille_regroupee'] = df['Famille_regroupee'].replace('Fruits et lÃ©gumes', 'Fruits et légumes')
```

```python
# Encodage de la variable 'Famille_regroupee' 

df = pd.concat([df, pd.get_dummies(df['Famille_regroupee'], prefix='Famille', drop_first=True)], axis=1)

```

```python
# Création du modèle de régression linéaire pour la charge glycemique (CG)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.metrics import mean_squared_error, r2_score

# 1. Identifier les colonnes booléennes et celles à exclure
bool_cols = df.select_dtypes(include='bool').columns.tolist()
cols_to_drop = ['Aliment', 'CG', 'IG'] + bool_cols

# 2. Définir X et y
X = df.drop(columns=cols_to_drop)
y = df['CG']

# 3. Préprocessing : encodage + standardisation
preprocessor = ColumnTransformer(
    transformers=[
        ('Encodage', OneHotEncoder(handle_unknown='ignore'), X.select_dtypes(include='object').columns),
        ('Standardisation', StandardScaler(), X.select_dtypes(include=['float64', 'int']).columns)
    ],
    remainder='drop'
)

# 4. Transformation des features
X_transformed = preprocessor.fit_transform(X)

# 5. Séparer en train/test
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

# 6. Régression linéaire avec contrainte de sortie positive
model = TransformedTargetRegressor(
    regressor=LinearRegression(),
    transformer=FunctionTransformer(np.log1p, inverse_func=np.expm1, validate=True)
)

# 7. Entraînement du modèle
model.fit(X_train, y_train)

# 8. Prédictions
y_pred = model.predict(X_test)

# 9. Évaluation
print("Erreur quadratique moyenne :", mean_squared_error(y_test, y_pred))
print("Score R² :", r2_score(y_test, y_pred))
```
